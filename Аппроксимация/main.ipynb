```python
from methods import *
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from funcs import *
import numpy as np
n=5
MAS1=[]
uzli=Randuzl(-1.5,1,60)
fuzli=values_generator(uzli)
x_vals = np.linspace(-1.5, 1, 100)
y_vals = [func(x) for x in x_vals]
plt.figure(figsize=(15, 25))  # Создание фигуры
for i in range(1,n+1):
    coefs=coefnormMNK(uzli,fuzli,i)
    y_approx=normMNK(x_vals, coefs)
    MAS1.append(lost(y_approx,y_vals))
    ax = plt.subplot(n, 1, i)
    ax.plot(x_vals, y_vals, label="Исходная функция", linewidth=1, color='blue')
    ax.plot(x_vals, y_approx, label=f"Аппроксимация полиномом степени {i}", linewidth=1, color='red')
    ax.scatter(uzli, fuzli, color='black', s=10, label="Узлы")  # Узлы как точки
    ax.legend()
    ax.grid()
    ax.set_title(f"Полином степени {i}")

plt.tight_layout()  # Корректируем расположение графиков
plt.show()


```
