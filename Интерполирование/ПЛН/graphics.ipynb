```python
from Newton import *
from Lagranzh import *
from funcs import *
import matplotlib.pyplot as plt
from prettytable import PrettyTable

table = PrettyTable()
cols=["Number of nodes","Number of test points","Max deviation RLn","Max deviation RLoptn"]
table.field_names=cols
mas=[]

fig,axes=plt.subplots(3,3,figsize=(25,25))
fig.suptitle("Интерполяция полиномом Ньютона")
a,b=-2,4
tests=[3,5,13,17,24,31,38,45,70]

for n in range(len(tests)):
    x_vals = np.linspace(a, b, 1000)
    XargsN = Nuzl(a, b, tests[n])
    XargsOpt = Optuzl(a,b,tests[n])
    YargsN = [f(x) for x in XargsN]
    YargsOpt=[f(x) for x in XargsOpt]
    y_true = [f(x) for x in x_vals]
    y_nuzl = [Nn(x, a,b, tests[n]) for x in x_vals]
    y_optuzl = [Noptn(x,a,b,tests[n]) for x in x_vals]

    mas.append(tests[n])
    mas.append(1000)
    mas.append(max_deviation(y_true,y_nuzl))
    mas.append(max_deviation(y_true,y_optuzl))
    table.add_row(mas)

    row, col = divmod(n, 3)  # Переводим индекс в (строка, столбец)
    ax = axes[row, col]

    # Рисуем графики
    ax.plot(x_vals, y_true, label="Исходная функция", linewidth=2)
    ax.plot(x_vals, y_nuzl, label=f"Интерполяция полиномом Ньютона - {tests[n]} равностоящих узлов", linestyle="--")
    ax.plot(x_vals, y_optuzl, label=f"Интерполяция полиномом Ньютона - {tests[n]} оптимальных узлов", linestyle="--")
    ax.scatter(XargsN, YargsN, color="red", label="Равностоящие узлы")
    ax.scatter(XargsOpt, YargsOpt, color="purple", label="Оптимальные узлы")

    ax.legend()
    ax.grid()
    ax.set_title(f"{tests[n]} узлов")

    mas=[]

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Корректируем расположение заголовка
plt.show()

```
